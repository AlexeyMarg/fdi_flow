# Example of using scalers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os
sys.path.append(os.path.abspath(".."))
from fdi_flow.preprocessing import scalers
```

## Generate test data

```python
n_points = 100
time = np.linspace(0, 10, n_points)
data1 = 3 * np.sin(time) + np.random.normal(0, 0.5, n_points)  
data2 = 0.5 * time + np.random.normal(0, 1, n_points)      

np_data = np.vstack([data1, data2])  # Форма (2, 100)

# Pandas DataFrame (100 точек × 2 ряда)
df_data = pd.DataFrame({
    'sensor1': data1,
    'sensor2': data2}, index=time)    
```

## Standard scaler

```python
scaler = scalers.StandardScaler()

np_scaled = scaler.apply(np_data, axis=0) 
df_scaled = scaler.apply(df_data)  

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(np_data[0], label='Original (row 0)')
plt.title('Numpy Array - Original (row 0)')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(np_scaled[0], label='Scaled (row 0)')
plt.title('Numpy Array - StandardScaled (row 0)')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(df_data.index, df_data['sensor1'], label='Original sensor1')
plt.title('DataFrame - Original')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(df_scaled.index, df_scaled['sensor1'], label='Scaled sensor1')
plt.title('DataFrame - StandardScaled')
plt.legend()
```

## Min max scaler

```python
scaler = scalers.MinMaxScaler()

np_scaled = scaler.apply(np_data, axis=0) 
df_scaled = scaler.apply(df_data)  

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(np_data[0], label='Original (row 0)')
plt.title('Numpy Array - Original (row 0)')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(np_scaled[0], label='Scaled (row 0)')
plt.title('Numpy Array - StandardScaled (row 0)')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(df_data.index, df_data['sensor1'], label='Original sensor1')
plt.title('DataFrame - Original')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(df_scaled.index, df_scaled['sensor1'], label='Scaled sensor1')
plt.title('DataFrame - StandardScaled')
plt.legend()
```

## Robust scaler

```python
scaler = scalers.RobustScaler()

np_scaled = scaler.apply(np_data, axis=0) 
df_scaled = scaler.apply(df_data)  

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(np_data[0], label='Original (row 0)')
plt.title('Numpy Array - Original (row 0)')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(np_scaled[0], label='Scaled (row 0)')
plt.title('Numpy Array - StandardScaled (row 0)')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(df_data.index, df_data['sensor1'], label='Original sensor1')
plt.title('DataFrame - Original')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(df_scaled.index, df_scaled['sensor1'], label='Scaled sensor1')
plt.title('DataFrame - StandardScaled')
plt.legend()
```
